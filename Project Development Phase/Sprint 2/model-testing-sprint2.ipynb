# Model Testing

>  ## Importing Packages :

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
model = load_model("../input/model-building-ibmnt/gesture.h5")
```

>## Loading The Image :

```python
img = image.load_img(r'../input/gesture-images/test/1/1.jpg', grayscale=True, target_size=(64,64))
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x,axis=0)
x.shape
```

> ## Predicting The Results :

```python
pred_x = model.predict(x)
pred_x=np.argmax(pred_x,axis=1)
pred_x
```

```python
index=['0', '1', '2', '3', '4', '5']
result=str(index[pred_x[0]])
result
```

```python
import numpy as np
p=[]
for i in range(0,6):
  for j in range(0,5):
    img = image.load_img(r"../input/gesture-images/test//"+str(i)+"//"+str(j)+".jpg", grayscale=True, target_size=(64,64))
    x = image.img_to_array(img)
    x = np.expand_dims(x,axis=0)
    pred = np.argmax(model.predict(x),axis=-1)
    p.append(pred)
print(p)
```

```python
result = []
index=['0','1','2','3','4','5']
for i in p:
  result.append(index[i[0]])
print(result)
```
